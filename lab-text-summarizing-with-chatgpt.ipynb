# Text summarizing with ChaptGPT
In this lesson, you will summarize text with a focus on specific topics.

## Setup

````python
from openai import OpenAI
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
````

````python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)


def get_completion(prompt, model="gpt-3.5-turbo"): # Andrew mentioned that the prompt/ completion paradigm is preferable for this class
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message.content
````

## Text to summarize

````python
prod_review = """
Got this panda plush toy for my daughter's birthday, \
who loves it and takes it everywhere. It's soft and \ 
super cute, and its face has a friendly look. It's \ 
a bit small for what I paid though. I think there \ 
might be other options that are bigger for the \ 
same price. It arrived a day earlier than expected, \ 
so I got to play with it myself before I gave it \ 
to her.
"""
````

## Summarize with a word/sentence/character limit

````python
prompt = f"""
Your task is to generate a short summary of a product \
review from an ecommerce site. 

Summarize the review below, delimited by triple 
backticks, in at most 30 words. 

Review: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

````python
prompt = f"""
Your task is to generate a short summary of a product \
review from an ecommerce site to give feedback to the \
Shipping deparmtment. 

Summarize the review below, delimited by triple 
backticks, in at most 30 words, and focusing on any aspects \
that mention shipping and delivery of the product. 

Review: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

````python
for i in range(len(reviews)):
    prompt = f"""
    Your task is to generate a short summary of a product \ 
    review from an ecommerce site. 

    Summarize the review below, delimited by triple \
    backticks in at most 20 words. 

    Review: ```{reviews[i]}```
    """

    response = get_completion(prompt)
    print(i, response, "\n")
````

````python
# Version 1: Structured Data Extraction (Sentiment, Pros, and Cons)
print("--- Version 1: Structured Pros & Cons ---")
for i, review in enumerate(reviews):
    prompt = f"""
    Analyze the product review delimited by triple backticks.
    Output your analysis strictly in this bulleted format:
    - Sentiment: (Positive, Neutral, or Negative)
    - Best Feature: (1 to 3 words max)
    - Biggest Flaw: (1 to 3 words max)

    Review: ```{review}```
    """
    response = get_completion(prompt)
    print(f"Product {i+1}:\n{response}\n")
````

````python
# Version 2: Department Action Items
print("--- Version 2: Action Item Generator ---")
for i, review in enumerate(reviews):
    prompt = f"""
    You are a customer success manager. Read the review delimited by triple backticks.
    Based on the review, what is the single most important action the company should take, 
    and which department (e.g., Shipping, QA, Pricing, Customer Service) is responsible?
    Keep your answer under 15 words.
    
    Review: ```{review}```
    """
    response = get_completion(prompt)
    print(f"Product {i+1}: {response}")
````

````python
# Version 3: The Social Media Manager
print("\n--- Version 3: The Honest Tweet ---")
for i, review in enumerate(reviews):
    prompt = f"""
    Summarize the product review delimited by triple backticks as a highly honest, 
    slightly sarcastic tweet. Include relevant emojis and hashtags. 
    You must keep the summary under 140 characters.
    
    Review: ```{review}```
    """
    response = get_completion(prompt)
    print(f"Product {i+1}: {response}")
````

I tested how GPT-3.5 handles summarizing and extracting data from a batch of four different product reviews (a toy, a lamp, a toothbrush, and a blender). Instead of just asking for a basic summary, I used a loop to run three creative prompts:

Extracting structured pros and cons.

Generating a specific action item for a company department.

Turning the review into an honest, slightly sarcastic tweet under 140 characters.

My findings
Overall, asking the model to extract specific information works much better than just asking it to "summarize." In version 1, it easily pulled the sentiment, best feature, and biggest flaw into clean bullet points for every single product. Version 2 was probably the most useful for a real business—it actually assigned logical tasks to the right teams, like telling the product development team to make the toothbrush head bigger, or telling the pricing department to stop jacking up the blender prices.

Where things broke (hallucinations and mistakes)
Looking closely at the outputs, I noticed two main places where the model struggled:

It cannot count: In version 3, I explicitly told the model to keep the tweets under 140 characters. The model completely failed at this. Almost every single tweet it generated was around 150 to 170 characters long. It understands the concept of "short," but it cannot do exact character math.

Forced negativity: In version 1, I forced the model to always list a "biggest flaw." The lamp review was actually super positive—a string broke in transit, but customer service fixed it immediately and the reviewer loved them. But because my prompt demanded a flaw, the AI extracted "string" as the biggest flaw, which makes the product sound worse than the reviewer intended.

What I learned

Extracting beats summarizing: If you just want useful data, ask the AI to extract specific bullet points. It cuts out all the useless fluff.

Character limits are useless: You can't trust an LLM to follow strict character or word counts. It's better to just give it a general limit (like "write one short sentence") instead of an exact number.

Be careful what you demand: If you force the AI to extract a negative aspect from a highly positive review, it will magnify a tiny detail just to fulfill your prompt. This could totally skew your data analytics if you aren't paying attention.
